```python
import requests
from bs4 import BeautifulSoup

base_url = 'https://ani.gamer.com.tw/animeList.php?sort=2&page='
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/96.0.4664.45 Safari/537.36',
}
start_page = 1
end_page = 5

anime_names = []
anime_watches = []

for page in range(start_page, end_page + 1):
    url = base_url + str(page)
    r = requests.get(url, headers=headers)

    if r.status_code == 200:
        print(f'請求成功：{r.status_code}')
        soup = BeautifulSoup(r.text, 'html.parser')
        anime_items = soup.select('.theme-list-block')

        # 依序針對每個動畫區塊擷取資料
        for anime_item in anime_items:
            anime_name_elements = anime_item.select('.theme-name')
            for element in anime_name_elements:
                anime_name = element.text.strip()
                anime_names.append(anime_name)

            anime_number_elements = anime_item.select('.theme-number:not(.theme-info-block)')
            for element in anime_number_elements:
                anime_watch = element.text.strip()
                anime_watches.append(anime_watch) 

            for i in range(len(anime_names)):
                print(anime_names[i])
                print(anime_watches[i])
                print('----------')

    else:
        print(f'請求失敗：{r.status_code}')

```

```python
import csv

# 創建一個空的列表來存儲數據
anime_data = []

# 將動畫名稱和觀看數放入數據列表中
for i in range(len(anime_names)):
    anime_data.append({'動畫名稱': anime_names[i], '集數': anime_watches[i]})

# 指定 CSV 檔案名稱
csv_filename = 'anime_data.csv'

# 開啟 CSV 檔案並寫入數據
with open(csv_filename, 'w', newline='', encoding='utf-8-sig') as csv_file:
    fieldnames = ['動畫名稱', '集數']
    writer = csv.DictWriter(csv_file, fieldnames=fieldnames)

    # 寫入表頭
    writer.writeheader()

    # 寫入數據
    for data in anime_data:
        writer.writerow(data)

print(f'數據已保存到 {csv_filename}')

```

```python
import json

# 數據
anime_data = []

# 創建 JSON 格式的數據
for i in range(len(anime_names)):
    anime_data.append({'動畫名稱': anime_names[i], '集數': anime_watches[i]})

# 指定 JSON 檔案名稱
json_filename = 'anime_data.json'

# 開啟 JSON 檔案並寫入數據
with open(json_filename, 'w', encoding='utf-8') as json_file:
    json.dump(anime_data, json_file, ensure_ascii=False, indent=4)

print(f'數據已保存到 {json_filename}')
```
